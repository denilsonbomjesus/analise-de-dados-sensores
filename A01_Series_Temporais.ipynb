#ACELEROMETRO - CAMINHADA LEVE (3 TESTES)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Acelerometro - Caminhada Leve

# Carregar os dados das tabelas CSV (substitua pelos nomes reais dos arquivos)
dt_acel_cleve1 = pd.read_excel('/content/data/sprint 1/acelerometro/1-caminhada_leve(testes)/Acelerao sem g 2023-08-19 17-24-16.xls')
dt_acel_cleve2 = pd.read_excel('/content/data/sprint 1/acelerometro/1-caminhada_leve(testes)/Acelerao sem g 2023-08-19 17-26-35.xls')
dt_acel_cleve3 = pd.read_excel('/content/data/sprint 1/acelerometro/1-caminhada_leve(testes)/Acelerao sem g 2023-08-19 17-28-33.xls')

# Selecionar os dados de interesse de cada tabela
time_acel_cleve1 = dt_acel_cleve1['Time (s)']
acceleration_x_acel_cleve1 = dt_acel_cleve1['Absolute acceleration (m/s^2)']

time_acel_cleve2 = dt_acel_cleve2['Time (s)']
acceleration_x_acel_cleve2 = dt_acel_cleve2['Absolute acceleration (m/s^2)']

time_acel_cleve3 = dt_acel_cleve3['Time (s)']
acceleration_x_acel_cleve3 = dt_acel_cleve3['Absolute acceleration (m/s^2)']

# Criar plot
plt.figure(figsize=(12, 6))

# Plot da primeira tabela
plt.plot(time_acel_cleve1, acceleration_x_acel_cleve1)
plt.xlabel('Time (s)')
plt.ylabel('Absolute Acceleration')
plt.title('Acceleration X Over Time - Table 1')

# Calcular estatísticas descritivas
mean_acceleration_x_acel_cleve1 = acceleration_x_acel_cleve1.mean()
std_deviation_acceleration_x_acel_cleve1 = acceleration_x_acel_cleve1.std()
dt_acel_cleve1_range_acceleration_x_acel_cleve1 = acceleration_x_acel_cleve1.max() - acceleration_x_acel_cleve1.min()
dt_acel_cleve1_range_acceleration_x_acel_cleve1 = acceleration_x_acel_cleve1.max() - acceleration_x_acel_cleve1.min()

# Criar o as linhas estatisticas
plt.plot(time_acel_cleve1, acceleration_x_acel_cleve1, label='Aceleração')
plt.axhline(y=mean_acceleration_x_acel_cleve1, color='r', linestyle='--', label='Média')
plt.axhline(y=std_deviation_acceleration_x_acel_cleve1, color='g', linestyle='--', label='Desvio Padrão')
plt.axhline(y=dt_acel_cleve1_range_acceleration_x_acel_cleve1 / 2, color='b', linestyle='--', label='Amplitude/2')
plt.legend()

# Exibir estatísticas no gráfico
stats_text = (
    f'Média: {mean_acceleration_x_acel_cleve1:.2f}\n'
    f'Desvio Padrão: {std_deviation_acceleration_x_acel_cleve1:.2f}\n'
    f'Amplitude: {dt_acel_cleve1_range_acceleration_x_acel_cleve1:.2f}'
)
print(stats_text)
```
